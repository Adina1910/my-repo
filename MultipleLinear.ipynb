```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
'''import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # for 3D plotting'''

# Create dataset
data = {
    'X1': [1, 2, 3, 4, 5],
    'X2': [2, 4, 6, 8, 10],
    'Y': [3, 6, 9, 12, 15]
}

df = pd.DataFrame(data)

# Independent and dependent variables
X = df[['X1', 'X2']]
y = df['Y']

# Train the model
model = LinearRegression()
model.fit(X, y)

# Print coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Predict for new data
new_data = pd.DataFrame([[6, 12]], columns=['X1', 'X2'])
prediction = model.predict(new_data)
print("Prediction for [6, 12]:", prediction[0])

''' ----- Visualization -----
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot actual points
ax.scatter(df['X1'], df['X2'], df['Y'], color='red', label='Actual Data', s=50)

# Create a grid for the regression plane
x1_grid, x2_grid = np.meshgrid(
    np.linspace(df['X1'].min(), df['X1'].max(), 10),
    np.linspace(df['X2'].min(), df['X2'].max(), 10)
)
y_pred_grid = model.intercept_ + model.coef_[0] * x1_grid + model.coef_[1] * x2_grid

# Plot regression plane
ax.plot_surface(x1_grid, x2_grid, y_pred_grid, alpha=0.5, color='skyblue', label='Regression Plane')

# Labeling
ax.set_xlabel('X1 (Hours Studied)')
ax.set_ylabel('X2 (Extra Practice)')
ax.set_zlabel('Y (Marks)')
ax.set_title('Multiple Linear Regression (3D Visualization)')
ax.legend()

plt.show()'''
```
