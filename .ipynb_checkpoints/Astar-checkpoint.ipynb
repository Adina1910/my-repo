```python
# Simple A* Search Algorithm

graph = {}
heuristic = {}

n = int(input("Enter number of nodes: "))
for i in range(n):
    node = input(f"\nEnter name of node {i+1}: ")
    neighbors = []
    k = int(input(f"Enter number of neighbors of {node}: "))
    for j in range(k):
        neigh = input(f"Enter neighbor {j+1} of {node}: ")
        cost = int(input(f"Enter cost from {node} to {neigh}: "))
        neighbors.append((neigh, cost))
    graph[node] = neighbors
    heuristic[node] = int(input(f"Enter heuristic value of {node}: "))

start = input("\nEnter starting node: ")
goal = input("Enter goal node: ")

open_list = [start]
closed_list = []
g_cost = {start: 0}
parent = {start: None}

while open_list:
    # --- Find node with smallest f = g + h ---
    current = open_list[0]
    for node in open_list:
        if g_cost[node] + heuristic[node] < g_cost[current] + heuristic[current]:
            current = node

    open_list.remove(current)
    closed_list.append(current)

    # Goal check
    if current == goal:
        path = []
        while current is not None:
            path.append(current)
            current = parent[current]
        path.reverse()
        print("\nPath found:", "-".join(path))
        break

    # Visit neighbors
    for (neighbor, cost) in graph[current]:
        if neighbor in closed_list:
            continue

        new_g = g_cost[current] + cost
        if neighbor not in open_list or new_g < g_cost.get(neighbor, float('inf')):
            g_cost[neighbor] = new_g
            parent[neighbor] = current
            if neighbor not in open_list:
                open_list.append(neighbor)
else:
    print("\nPath does not exist!")
```
