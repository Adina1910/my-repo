```python
# Greedy Best First Search (GBFS)

# Create an empty graph (dictionary)
graph = {}

# Take user input for number of nodes
n = int(input("Enter number of nodes: "))

# Take input for graph connections
for i in range(n):
    node = input(f"Enter node name {i+1}: ")
    neighbors = input(f"Enter neighbors of {node} (separated by space): ").split()
    graph[node] = neighbors

# Take heuristic values for each node
heuristic = {}
print("\nEnter heuristic values for each node:")
for node in graph:
    h = int(input(f"Heuristic value of {node}: "))
    heuristic[node] = h

# Take starting node and goal node
start = input("\nEnter starting node: ")
goal = input("Enter goal node: ")

# GBFS algorithm
visited = []
queue = [start]

print("\nGBFS Traversal:")

while queue:
    # --- find the node with smallest heuristic manually ---
    smallest = queue[0]
    for node in queue:
        if heuristic[node] < heuristic[smallest]:
            smallest = node

    # remove the node with smallest heuristic
    queue.remove(smallest)

    if smallest not in visited:
        print(smallest, end=" ")
        visited.append(smallest)

        # stop if goal reached
        if smallest == goal:
            print("\nGoal reached!")
            break

        # add unvisited neighbors
        for neighbor in graph[smallest]:
            if neighbor not in visited:
                queue.append(neighbor)
```
