```python
#
# Simple Data Analysis Demo - Focuses on the core three libraries:
# NumPy (Creation), Pandas (Organization), and Matplotlib (Visualization).
#
# NOTE: Error checking (try/except blocks) has been removed to simplify the code.
# The script assumes valid numeric input will be provided by the user.
#

# --- 1. Import Core Libraries ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- Function to Get User Input for Data ---
def get_user_data():
    """Collects study hours and quiz scores from the user using simple input."""
    
    hours_list = []
    score_list = []
    
    # Get the number of students (assuming user enters a valid integer)
    num_students = int(input("Enter the number of students (data points) you want to enter: "))
    
    print(f"\n--- Collecting data for {num_students} students ---")

    # Loop to collect data for each student
    for i in range(num_students):
        print(f"\n--- Student {i + 1} of {num_students} ---")
        
        # Get Hours Studied (assuming user enters a valid number/float)
        hours = float(input(f"Enter Hours Studied for Student {i + 1} (e.g., 4.5): "))
        hours_list.append(hours)
        
        # Get Quiz Score (assuming user enters a valid number/float)
        score = float(input(f"Enter Quiz Score (0-100) for Student {i + 1} (e.g., 85): "))
        score_list.append(score)

    # Create the data dictionary and DataFrame components
    data = {
        'Student_ID': np.arange(1, num_students + 1),
        'Hours_Studied': hours_list,
        'Quiz_Score': score_list
    }
    
    return data

# --- MAIN EXECUTION ---

# Get data from user
data = get_user_data()

# Check if data was successfully collected before processing
if not data['Hours_Studied']:
    print("No data entered. Exiting script.")
else:
    print("\n--- Data Creation and Organization (NumPy & Pandas) ---")
    
    # Pandas: Create a DataFrame (a structured table)
    df = pd.DataFrame(data)

    print("\nStructured Data (DataFrame Head):")
    print(df.head())

    # --- 2. Explicit Data Manipulation (Pandas) ---

    # Manipulation Example 1: Creating a new column by calculation
    # Calculate 'Points_Per_Hour' to see the efficiency of studying
    df['Points_Per_Hour'] = df['Quiz_Score'] / df['Hours_Studied']

    # Manipulation Example 2: Filtering data
    # Get only the students who scored 85 or higher
    high_scorers = df[df['Quiz_Score'] >= 85]
    low_scorers = df[df['Quiz_Score'] < 85]
    
    print("\nData after new column calculation:")
    print(df.head())

    print("\nFiltered Data (High Scorers):")
    print(high_scorers)
    print("\nFiltered Data (Low Scorers):")
    print(low_scorers)

    print("\nSummary Statistics for all students:")
    # Get quick insights into the numerical data (mean, min, max, etc.)
    print(df.describe())

    # --- 3. Data Visualization (Matplotlib) ---
    print("\n--- Data Visualization (Matplotlib) ---")
    
    # Matplotlib: Create a single, clear scatter plot
    plt.figure(figsize=(8, 6))

    # Plotting the relationship between Hours Studied (X) and Quiz Score (Y)
    plt.scatter(df['Hours_Studied'], df['Quiz_Score'], color='purple', alpha=0.8)

    plt.title('Relationship between Study Hours and Quiz Score (User Data)')
    plt.xlabel('Hours Studied')
    plt.ylabel('Quiz Score')
    plt.grid(True, linestyle='--')

    # Display the plot
    plt.show()

    print("\nScript completed. A scatter plot was generated using your input data.")
```
