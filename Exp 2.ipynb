```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("--- Data Creation and Organization (NumPy & Pandas) ---")

# NumPy: Use simple lists/arrays for data creation
data = {
    'Student_ID': np.arange(1, 11),
    # Student hours studied
    'Hours_Studied': [2.5, 3.0, 4.2, 5.5, 6.0, 1.5, 7.8, 4.0, 5.0, 6.5],
    # Corresponding quiz scores
    'Quiz_Score': [70, 75, 85, 90, 95, 60, 98, 80, 88, 92]
}

# Pandas: Create a DataFrame (a structured table)
df = pd.DataFrame(data)

print("\nStructured Data (DataFrame Head):")
print(df.head())

# --- 2. Explicit Data Manipulation (Pandas) ---

# Manipulation Example 1: Creating a new column by calculation
# Calculate 'Points_Per_Hour' to see the efficiency of studying
df['Points_Per_Hour'] = df['Quiz_Score'] / df['Hours_Studied']

# Manipulation Example 2: Filtering data
# Get only the students who scored 85 or higher
high_scorers = df[df['Quiz_Score'] >= 85]
low_scorers = df[df['Quiz_Score']<85]
print("\nData after new column calculation:")
print(df.head())

print("\nFiltered Data (High Scorers):")
print(high_scorers)
print("\nFiltered Data (Low Scorers):")
print(low_scorers)

print("\nSummary Statistics for all students:")
# Get quick insights into the numerical data (mean, min, max, etc.)
print(df.describe())

print("\n--- Data Visualization (Matplotlib) ---")

# Matplotlib: Create a single, clear scatter plot
plt.figure(figsize=(8, 6))

# Plotting the relationship between Hours Studied (X) and Quiz Score (Y)
plt.scatter(df['Hours_Studied'], df['Quiz_Score'], color='purple', alpha=0.8)

plt.title('Relationship between Study Hours and Quiz Score')
plt.xlabel('Hours Studied')
plt.ylabel('Quiz Score')
plt.grid(True, linestyle='--')

# Display the plot
plt.show()
```
