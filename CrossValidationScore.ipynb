```python
import pandas as pd
from sklearn.model_selection import cross_val_score, KFold
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Load dataset from your CSV file
data = pd.read_csv("iris_data.csv")

# Separate features and target column
X = data.drop("class", axis=1)
y = data["class"]

# Define model
model = KNeighborsClassifier()

# Define number of folds
folds = 5
kfold = KFold(n_splits=folds, shuffle=True, random_state=1)

# Perform cross-validation
scores = cross_val_score(model, X, y, cv=kfold)

# Display results
print("--- Cross Validation Results ---")
print("Model Used: K-Nearest Neighbors (KNN)")
print("Dataset: iris_data.csv")
print("Number of Folds:", folds)
print("Scores for each fold:", scores)
print("Average Accuracy:", round(np.mean(scores), 4))
print("Standard Deviation:", round(np.std(scores), 4))
```
