```python
# Create an empty graph (dictionary)
graph = {}

# Take user input for number of nodes
n = int(input("Enter number of nodes: "))

# Take input for graph connections
for i in range(n):
    node = input(f"Enter node name {i+1}: ")
    neighbors = input(f"Enter neighbors of {node} (separated by space): ").split()
    graph[node] = neighbors

# Take starting node
start = input("Enter starting node for DFS: ")

# DFS algorithm using stack (list)
visited = []
stack = [start]  # stack to keep track of nodes to visit

print("\nDFS Traversal:")

while stack:
    node = stack.pop()  # remove last element (LIFO)
    if node not in visited:
        print(node, end=" ")
        visited.append(node)
        # Add neighbors to stack (in reverse order to maintain order)
        for neighbor in graph[node]:
            if neighbor not in visited:
                stack.append(neighbor)
```
